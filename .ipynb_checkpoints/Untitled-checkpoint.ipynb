```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cell 1: Read the data
print("Cell 1: Loading Data")
print("-" * 50)
df = pd.read_csv(r'C:\Users\RTX\Desktop\newia\emotions\Emotion_classify_Data.csv')
print(f"Total number of sentences: {len(df)}")
print("\nFirst few rows of the dataset:")
print(df.head())
print("\n")

# Cell 2: Count emotions
print("Cell 2: Emotion Distribution")
print("-" * 50)
emotion_counts = df['Emotion'].value_counts()
print("Number of sentences for each Emotion:")
print(emotion_counts)
print("\n")

# Cell 3: Visualization of Emotion distribution
print("Cell 3: Visualizing Emotion Distribution")
print("-" * 50)
plt.figure(figsize=(10, 6))
emotion_counts.plot(kind='bar')
plt.title('Distribution of Emotions in Dataset')
plt.xlabel('Emotion')
plt.ylabel('Number of Sentences')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("\n")

# Cell 4: Analyze sentence lengths
print("Cell 4: Sentence Length Analysis")
print("-" * 50)
df['word_count'] = df['sentence'].str.split().str.len()
print("Average words per sentence for each Emotion:")
print(df.groupby('Emotion')['word_count'].mean())
print("\nMaximum words per sentence for each Emotion:")
print(df.groupby('Emotion')['word_count'].max())
print("\nMinimum words per sentence for each Emotion:")
print(df.groupby('Emotion')['word_count'].min())
print("\n")

# Cell 5: Common words by Emotion
print("Cell 5: Common Words Analysis")
print("-" * 50)
for Emotion in df['Emotion'].unique():
    print(f"\nMost common words in {Emotion} sentences:")
    # Get all words from sentences of this Emotion
    words = ' '.join(df[df['Emotion'] == Emotion]['sentence']).lower().split()
    # Remove basic punctuation
    words = [word.strip('.,!?') for word in words]
    # Count word frequencies
    word_freq = pd.Series(words).value_counts().head(5)
    print(word_freq)

# Cell 6: Analyze punctuation
print("\nCell 6: Punctuation Analysis")
print("-" * 50)
df['exclamation_count'] = df['sentence'].str.count('!')
print("\nAverage exclamation marks per Emotion:")
print(df.groupby('Emotion')['exclamation_count'].mean())

# Cell 7: Save summary to file
print("\nCell 7: Saving Summary")
print("-" * 50)
summary = pd.DataFrame({
    'Total_Sentences': emotion_counts,
    'Avg_Word_Count': df.groupby('Emotion')['word_count'].mean(),
    'Avg_Exclamations': df.groupby('Emotion')['exclamation_count'].mean()
})
print("\nFinal Summary:")
print(summary)

# Optionally save to CSV
summary.to_csv('emotion_analysis_summary.csv')
print("\nSummary saved to 'emotion_analysis_summary.csv'")
```
